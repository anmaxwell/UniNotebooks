```
import pandas as pd
import numpy as np


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier 
from spacy.lang.en import English
```

```
df = pd.read_csv("amazon_cells_labelled.txt", sep='\t', header = None,
index_col=False, names=['review', 'label'],)
```

```
nlp = English()
text_data = text_data = np.array(df.iloc[:, 0])
```

```
# Create a list of the unique headings
token_list = []
for row in text_data:
    text = nlp(row)
    for token in text: 
        token_list.append(token.text)

unique_tokens = set(token_list)
```

```
# Create the bag of words feature matrix
count = CountVectorizer()
bag_of_words = count.fit_transform(text_data)

# Show feature matrix
bag_of_words.toarray()
```

```
type(bag_of_words)
```

```
dataset = pd.DataFrame(bag_of_words.toarray(), columns=count.get_feature_names())
```

```
dataset['label'] = df.iloc[:, 1]
dataset.head()
```

```
#Split the data into a training and a testing set
train_features = dataset.iloc[:800,:-1]
test_features = dataset.iloc[800:,:-1]
train_labels = dataset.iloc[:800,-1]
test_labels = dataset.iloc[800:,-1]
```

```
tree = DecisionTreeClassifier(criterion = 'entropy').fit(train_features,train_labels)
```

```
#The accuracy is then calculated through the score function offered by sklearn
print("The prediction accuracy is: ",tree.score(test_features,test_labels)*100,"%")
```
